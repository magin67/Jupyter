```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread("Dm.png", format=None)
plt.imshow(img);
#img.shape
```

```python
def extremeCoord(mVal, numPoints=5, min=True):
    CoordX, CoordY, Val = np.zeros(numPoints), np.zeros(numPoints), np.zeros(numPoints)
    for i in range(numPoints):
        if min:
            extind = np.argmin(mVal)
        else:
            extind = np.argmax(mVal)
        index = np.unravel_index(extind, mVal.shape)
        row, col = index[0], index[1]
        val = mVal[row, col]
        CoordX[i], CoordY[i], Val[i] = row, col, val
        if min:
            mVal[row, col] = 1
        else:
            mVal[row, col] = 0

    return (CoordX, CoordY, Val)

def ToLevels(mVal, numLevels=10):
    return np.rint(mVal*numLevels)/numLevels

imgR = img[:, :, 0]
imgG = img[:, :, 1]
imgB = img[:, :, 2]

numL = 3
imgA = ToLevels((imgR + imgG + imgB)/3, numL)
#imgL = ToLevels(imgA, 2)
#rec = extremeCoord(imgA, 1000, True)
plt.imshow(imgA, cmap="binary_r"); #"binary_r""Dark2"
#plt.scatter(rec[1], rec[0], s=10);
#print(rec[0])
```
