Исследуем применимость параметров симплекса для анализа данных электроизмерений на грунте.
Используем данные измерений Кушман, 2011.

Симплекс - это набор вершин с известными дистанциями между ними.
Известно, что матрица дистанция симплекса позволяет определить его геометрические параметры.

Параметры симплекса можно разделить на две группы.
К первой относятся параметры, связанные с его размером. К ним можно отнести:
- объем симплекса: vol
- среднюю дистанцию в симплексе: ra
- радиус описанной сферы (внешний): rs
- внутренний радиус: ri

Все дистанцияя здесь - это квадраты расстояния.

Вторая группа параметров - это безразмерные параметры - индексы, которые вчисляются как отношение размерных, так чтобы размерности сократились.
К индексам относятся:
- индекс симметричности, - отношение средней дистанции к внешнему радиусу: ra/rs
- индекс радиусов, - отношение внутреннего радиуса к внешнему: ri/rs
- индекс объема, - отношение объема к степени внешнего радиуса: vol^2/rs^(n-1)

Индексы позволяют сравнивать между собой значения для различных размеров симплекса, поэтому удобны при построении вертикальных срезов.

В процессе ислледования собираемся в частности выяснить, какие параметры имеют самостоятельное значение, а какие нет.

Набор вершин симплекса можно формировать по разному. 
Зесь используем квадратную структуру (установку), вершины симплекса - это вершины квадрата.
Меняем значение размаха - от 1 до 3. Схема расположения вершин:

- o  o
- o  o   

При большом количестве вершин симплекса значительно повышается чувствительность к точности измерений. При 16 вершинах обращение матриц становится нестабильным - большое влияние погрешностей.

Подготовим процедуры для анализа данных.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

import numpy as np
from numpy import linalg as la

import NetGeometry as ng
import ioData as io

colors = [(0, 0, 1), (0, 1, 0), (1, 0, 0)]  # B -> G -> R
cm = LinearSegmentedColormap.from_list('my_list', colors, N=10)
```

```python
lData = io.File2Data('Данные (Кушман, 2011).dat') # загрузка данных
mData, numNodes = io.Data2Array(lData)

def ScanData(NumRow, NumCol, size, idParam='rs', degree=1):
    # прогонка установки по массиву данных
    mNodes = io.mNodesCoordinate(NumRow, NumCol) # координаты узлов
    mres = np.zeros((NumRow-size, NumCol-size)) # результаты
    maska = io.SquareMaska(size) # задает вид установки

    mIdent = np.eye(4)
    for row in range(0, NumRow-size):
        for col in range(0, NumCol-size):
            baseSet = io.SetOfNodes(mNodes, maska, row, col)
            mDistance = io.mMinor(mData, baseSet)
            if degree != 1:
                mDistance += mIdent
                mDistance = mDistance**(degree) - mIdent
            pbData = ng.PointBase(mDistance)
            mres[row,col] = pbData.value(idParam)
            # if (mres[row,col] < 0): print(row, col, size, mres[row,col])
            
    return mres

def ShowPlots(idParam='', contour=0, low=0, up=0):
    # вывод данных на график
    fig, axs = plt.subplots(1, 3, figsize=(10, 3))
    for i in range(3):
        mRs = ScanData(19, 13, i+2, idParam)
        ax = axs[i]
        im = ax.imshow(mRs, cmap=cm, interpolation='bicubic');
        if low*up > 0:
            im.set_clim(low, up)
        if contour > 0:
            ax.contour(mRs, contour, linewidths=1) # , extent=(-3, 3, -2, 2)
        # origin='lower',
        # interpolation:
        #   ‘none’, ‘nearest’, ‘bilinear’, ‘bicubic’, ‘spline16’, ‘spline36’, ‘hanning’, ‘hamming’, ‘hermite’,
        #   ‘kaiser’, ‘quadric’, ‘catrom’, ‘gaussian’, ‘bessel’, ‘mitchell’, ‘sinc’, ‘lanczos’
        ax.set_title('Spacing: '+ str(i+1))
        ax.grid(True)
        fig.colorbar(im, ax=ax)
```

### Первая группа параметров - размерные

```python
ShowPlots('ra', 10) # Средняя дистанция
```

```python
# ShowPlots('rs', 10) # Внешний радиус - не отличается от ra
```

```python
ShowPlots('rh', 10) # Внутренний радиус
```

```python
ShowPlots('sq2', 10) # Сумма квадратов площадей граней
```

```python
ShowPlots('vol', 10) # Объем
```

Видим, что картинка на всех размерных параметрах примерно одинакова.
То есть данные сильно коррелируют между собой.
